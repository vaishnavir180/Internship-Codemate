```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#scatter plot
```

```python
x=np.arange(0,10)
y=np.arange(11,21)
```

```python
print(x)
print(y)
```

```python
plt.scatter(x,y,c='g')
plt.xlabel ('x axis')
plt.ylabel ('y axis')
plt.title('Graph in 2D')
plt.savefig('Test.png')
plt.show()
```

```python
#plot
```

```python
y=x*x
```

```python
plt.figure(figsize=(8,4))
plt.plot(x,y,'go--',linewidth=2)
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('2d Digram')
plt.show()
```

```python
#subplot
```

```python
plt.subplot(2,2,1)
plt.plot(x,y,'r--')

plt.subplot(2,2,2)
plt.plot(x,y+1,'g*--')

plt.subplot(2,2,3)
plt.plot(x,y*2,'bo--')

plt.subplot(2,2,4)
plt.plot(x,y,'go--')

plt.show()
```

```python
#plot  generating equation
```

```python
x = np.arange(1,11)
y = 3 * x + 5
plt.title("Matplotlib demo")
plt.xlabel("x axis caption")
plt.ylabel("y axis caption")
plt.plot(x,y)
plt.show()
```

```python
np.pi
```

```python
plt.figure(figsize=(6,3))
x = np.arange(0,4  * np.pi,0.1)
y = np.sin(x)
plt.title=("sine wave form")
plt.plot(x,y)
plt.show()
```

```python
#bar plot
```

```python
x1=[1,2,3]
y1=[4,5,6]

x2=[8,9,10]
y2=[4,8,6]

x3=[3,6,9]
y3=[2,4,6]

plt.bar(x1,y1,color="g")
plt.bar(x2,y2,color="b")
plt.bar(x3,y3,color="r")

plt.title("bar graph")
plt.ylabel("y axis")
plt.xlabel("x axis")

plt.show()
```

```python
#histogram
```

```python
height = np.random.randint(150,180,40)
plt.hist(height,orientation ='vertical',color='c')
plt.title("histogram")
plt.show()
```

```python
#box plot
```

```python
sample = np.array([15,111,18,150,13,17,11,22,5,18,10])
sample
```

```python
plt.figure(figsize=(6,4))
plt.boxplot(sample)
plt.title(" Detecting outliers using Boxplot")
plt.xlabel('sample')
```

```python
#pie chart
```

```python
labels=['python','c++','java','R']
x=[50,50,60,70]
colors=['gold','yellowgreen','lightcoral','lightskyblue']
explode=(0,0.1,0,0.1)

plt.pie(x,explode=explode ,labels=labels,colors=colors,autopct='%1.1f%%',shadow='true')
plt.show()
```
