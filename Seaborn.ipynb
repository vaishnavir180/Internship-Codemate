```python
import seaborn as sns
import pandas as pd 

import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('diabetes.csv')
```

```python
df.head()
```

```python
import seaborn as sns 
df.head()
df.corr()  #correlation value between 1 and -1
```

```python
#heatmap
sns.heatmap(df.corr(),annot=True,cmap='bwr')
```

```python
sns.heatmap(df.corr(),annot=False,cmap='ocean')
```

```python
#Generate Joinplot for age vs pregnancy.
sns.jointplot(x='Age',y='Pregnancies',data=df,kind='hex')
#plot_kinds = ["scatter", "hist", "hex", "kde", "reg", "resid"]
```

```python
df.columns
```

```python
#pair plot
sns.pairplot(df[['Glucose', 'BloodPressure', 'SkinThickness','Age', 'Outcome']],hue='Outcome')
```

```python
#distribution plot
sns.distplot(df['Age'])
```

```python
sns.distplot(df['Glucose'])
```

```python
sns.distplot(df['Insulin'])
```

```python
#count plot
sns.countplot(df['Outcome'])
```

```python
sns.countplot(x=df['Outcome'],palette='rainbow')
```

```python
#bar plot
import matplotlib.pyplot as plt
plt.figure(figsize=(12,4))

sns.barplot(x='Age',y='Outcome',data=df)
```

```python
#box plot
sns.boxplot(x='Age',data=df)
```

```python
#violin plot
sns.violinplot(x=df['Age'],palette='rainbow')
# shift tab
```
